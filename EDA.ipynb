# IPS Column for feature Engineering

```python
# feature engineering apply again it means we add a column for know about  them into two value (0,1) for checkng true or false 
data['Ips'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data.head()
```

```python
new = data['ScreenResolution'].str.split('x')
new
```

```python
# we fetch the value from screenresolution put them into in two new column x and y
new = data['ScreenResolution'].str.split('x',n=1,expand=True)
new
```

```python
data['x_res'] = new[0] 
data['y_res'] = new[1]
```

```python
data.head()
```

```python
# here create some problem now we can write some Regular expression, x not show in the form of int
data['x_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)')
```

```python
data['x_res'] = data['x_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
```

```python
data.head()
```

```python
#convert object into integer
data['x_res'] = data['x_res'].astype('int32')
data['y_res'] = data['y_res'].astype('int32')
```

```python
data['ppi'] = (((data['x_res']**2) + (data['x_res']**2))**0.5/data['Inches']).astype('float')
```

```python
data['ppi']
```

```python
data.head()
```

```python
#Droping the Column ScreenResolution,X,y,ips
```
